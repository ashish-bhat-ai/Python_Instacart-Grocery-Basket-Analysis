# 02. Wrangling data

## Dropping unneeded columns and reseting the df

```python
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
df_ords.head()
```

```python
# using value counts to find missing values info
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

```python
# renaming a column to more intuitive one
df_ords.rename(columns = {'order_dow':'orders_day_of_the_week'},inplace=True)
```

```python
df_ords.head()
```

## Changing data type of a column

```python
# Changing data type to string/object in order for describe function to not summarize it
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['order_id'].dtype
```

## Transposing data ( rows <> columns )

```python
# Importing new data set departments
df_dep = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'departments.csv'), index_col = False)
```

```python
df_dep.head()
```

```python
# trabsposing and storing in a new df
df_dep_t = df_dep.T
df_dep_t
```

```python
# Resetting index to take care of the weird 0 on original index
df_dep_t.reset_index()
```

```python
# Storing original header in a new variable to use it later
new_header = df_dep_t.iloc[0]
```

```python
# create a new dataframe that only copies over rows beyond the first row
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# Use the stored original headers to assign column headers df.columns = stored values
df_dep_t_new.columns = new_header
df_dep_t_new
```

## Data dictionaries

```python
# Converting a file with ids and names to a data dictionary
data_dict = df_dep_t_new.to_dict('index') # tf.to_dict - function and index tells python to use numbered rows as key values in functions
data_dict
```

```python
df_prods.head()
```

```python
# How to use data dictionary to see whats in the value
print(data_dict.get('19'))
```

## Subsetting : Creating a smaller data set from a bigger set based on particular filters

```python
# important for this - indexing or retrieving an element - object['element']
```

```python
# df_prods['department_id']==19 - creates a column with value as true when department_id=19, otherwise false
df_prods['department_id']==19
```

```python
# putting this indexing inside df_prods[] tells to retrive only values in the column which are true
df_snacks =  df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# ANOTHER WAY TO DO THIS
```

```python
# USING loc function : while iloc looks for an integer position, loc looks for a column name
```

```python
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
# If we want to look for multiple values , we use isin([  ]) function
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```

## Exporting the processed data frame to the folders

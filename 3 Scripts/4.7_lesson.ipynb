# B. Defining the price label using if,elif,else - user defined function

```python
# Defining the price label function - user generated

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying the function to store values returned in the new column
df['price_range'] = df.apply(price_label, axis=1)
```
